```python
from docx import Document
import pandas as pd

# Đọc file Word
doc = Document('git.docx')

# Khởi tạo danh sách để lưu dữ liệu
data = []
current_command = None

# Duyệt qua tất cả các đoạn văn trong tài liệu Word
for para in doc.paragraphs:
    text = para.text.strip()

    # Kiểm tra xem đoạn văn có phải là một lệnh git (được nhận dạng bằng việc chứa '[1]', '[5]', '[7]')
    if '[1]' in text or '[5]' in text or '[7]' in text:
        # Nếu có lệnh hiện tại và mô tả tương ứng, thêm vào danh sách
        if current_command is not None:
            data.append([current_command, description.strip()])

        # Bắt đầu một lệnh mới
        current_command = text
        description = ""
    else:
        # Nếu đoạn văn không phải là lệnh git, thêm nó vào mô tả của lệnh hiện tại
        if current_command:
            description += text + " "

# Thêm lệnh cuối cùng vào danh sách
if current_command is not None:
    data.append([current_command, description.strip()])

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data, columns=['Command', 'Description'])
df
```

```python
# Sử dụng phương thức str.replace để loại bỏ các ký tự '[1]', '[5]', '[7]'
df['Command'] = df['Command'].str.replace(r'\[\d\]', '', regex=True)

df
```

```python
# Lưu DataFrame dưới dạng bảng trong file txt
df.to_csv('git_commands.txt', sep='\t', index=False)

print("Dữ liệu đã được lưu thành file git_commands.txt")
```

Xử lý phần conda cmd

```python
from docx import Document

# Đọc file Word
doc = Document('conda.docx')  # Đảm bảo rằng file có định dạng .docx

# Khởi tạo danh sách để lưu dữ liệu
lines = []

# Đọc tất cả các đoạn văn từ file doc và thêm vào danh sách lines
for para in doc.paragraphs:
    lines.append(para.text.strip())

# Khởi tạo danh sách để lưu trữ kết quả
new_lines = []
current_conda_command = ""

# Duyệt qua từng dòng
for line in lines:
    stripped_line = line.strip()
    
    # Nếu dòng chứa từ "Conda", lưu lại dòng đó là lệnh conda hiện tại
    if "Conda" in stripped_line:
        current_conda_command = stripped_line
    
    # Nếu dòng bắt đầu bằng "-", thay thế dấu phẩy bằng '/', thêm lệnh conda hiện tại vào đầu dòng và thêm '[]' ở cuối dòng
    if stripped_line.startswith('-'):
        modified_line = stripped_line.replace(',', '/')
        new_lines.append(current_conda_command + " " + modified_line)
    else:
        new_lines.append(stripped_line)

# Ghép các dòng lại thành chuỗi văn bản mới
new_text = '\n'.join(new_lines)

# In ra kết quả hoặc lưu vào file
print(new_text)
```

```python
# Lưu vào file .txt
with open('conda_commands.txt', 'w', encoding='utf-8') as f:
    f.write(new_text)
```

```python
import pandas as pd

# Đọc nội dung file văn bản
with open('conda_commands.txt', 'r') as file:
    lines = file.readlines()

# Khởi tạo danh sách để lưu dữ liệu
data = []
current_command = None
description = []

# Xử lý các dòng
for line in lines:
    line = line.strip()
    if 'Conda' in line:
        # Nếu có lệnh hiện tại, thêm vào danh sách
        if current_command is not None:
            data.append([current_command, '\n'.join(description).strip()])

        # Bắt đầu một lệnh mới
        current_command = line
        description = []
    else:
        # Nếu không phải lệnh, thêm vào mô tả của lệnh hiện tại
        if current_command:
            description.append(line)

# Thêm lệnh cuối cùng vào danh sách
if current_command is not None:
    data.append([current_command, '\n'.join(description).strip()])

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data, columns=['Command', 'Description'])

df
```

```python
# Thêm df1 vào file git_commands_cleaned.txt
df.to_csv('git_commands.txt', sep='\t', index=False, header=False, mode='a')

print("Dữ liệu từ df1 đã được thêm vào file git_commands.txt")
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.interviewbit.com/blog/sql-commands/'

try:
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the div (check if it exists before accessing methods)
        div_element = soup.find('div', class_='entry-content')

        if div_element:
            # Find all ul elements
            ul_elements = div_element.find_all('ul')

            # Initialize empty lists for results
            titles = []
            contents = []

            for ul_element in ul_elements:
                # Extract title from li elements within ul
                li_elements = ul_element.find_all('li')
                title_text = ''.join([li.text.strip() for li in li_elements])  # Combine text from all li within ul

                # Find the next sibling p element (assuming it's directly below)
                next_p = ul_element.find_next_sibling('p')
                content_text = next_p.text.strip() if next_p else None  # Handle cases where p might not exist

                # Append results
                titles.append(title_text)
                contents.append(content_text)

            # Print results
            for i in range(len(titles)):
                print(titles[i])
                print(contents[i])  # Add newline for better formatting
        else:
            print("Div element with class 'entry-content' not found.")

    else:
        print(f"Failed to retrieve webpage. Status code: {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"Error occurred during request: {e}")
```

```python
new_titles = titles[4:-2]
new_contents = contents[4:-2]

# Lưu kết quả vào file
with open("SQL_commands.txt", "w", encoding="utf-8") as f:
    for i in range(len(new_titles)):
        f.write(new_titles[i] + "\n")  # Thêm xuống dòng sau mỗi title
        f.write(new_contents[i] + "\n")  # Thêm hai dòng trống sau mỗi content để phân cách dễ nhìn hơn
```

```python
url = 'https://www.tecmint.com/dangerous-linux-commands/'

try:
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # tìm tất cả thẻ div entry-content
        div_element = soup.find('div', class_='entry-content')

        # Tìm tất cả các thẻ <h3> chứa tiêu đề
        cmds = div_element.find_all('h3')

        # In nội dung trong mỗi thẻ <h3>
        for cmd in cmds:
            print(cmd.text)  # In ra nội dung của thẻ h3
    else:
        print(f"Failed to retrieve webpage. Status code: {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"Error occurred during request: {e}")
```
