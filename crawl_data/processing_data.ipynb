Python commands

```python
import pandas as pd
```

```python


# Đọc nội dung từ file txt
with open("python_commands.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Khởi tạo danh sách cho hai cột
cmd_list = []
discrip_list = []

# Duyệt qua các dòng và tìm các dòng bắt đầu bằng 'Syntax:'
for i in range(len(lines)):
    if lines[i].startswith("Syntax:"):
        cmd_list.append(lines[i].strip())  # Lấy dòng Syntax
        if i + 1 < len(lines):  # Kiểm tra xem có dòng sau dòng Syntax không
            discrip_list.append(lines[i + 1].strip())  # Lấy dòng Example

# Tạo DataFrame với hai cột
df = pd.DataFrame({
    "Commands": cmd_list,
    "Discription": discrip_list
})

df
```

```python
# Xóa chữ 'Syntax:' khỏi cột Syntax
df['Commands'] = df['Commands'].str.replace("Syntax:", "").str.strip()
df
```

SQL commands

```python
import pandas as pd

# Giả sử file SQL_commands.txt có nội dung như mô tả
with open("SQL_commands.txt", "r", encoding="utf-8") as file:
    lines = file.read()

# Tách nội dung dựa trên từng lệnh
commands = [
    "CREATE Command:", "DROP Command:", "ALTER Command:", "TRUNCATE Command:", "INSERT Command:",
    "UPDATE Command:", "DELETE Command:", "GRANT Command:", "REVOKE Command:",
    "COMMIT Command:", "ROLLBACK Command:", "SAVEPOINT Command:"
]

# Khởi tạo danh sách cho các cột mới
command_list = []
description_list = []

# Duyệt qua từng lệnh và tách lệnh với phần mô tả tương ứng
for i in range(len(commands)):
    start_idx = lines.find(commands[i])
    if i + 1 < len(commands):
        end_idx = lines.find(commands[i + 1])
    else:
        end_idx = len(lines)
    
    command_list.append(commands[i].replace(" Command:", ""))  # Lưu lệnh, bỏ chữ "Command:"
    description_list.append(lines[start_idx + len(commands[i]):end_idx].strip())  # Lưu phần mô tả

# Tạo DataFrame mới với hai cột Command và Description
df1 = pd.DataFrame({
    "Commands": command_list,
    "Discription": description_list
})

df1
```

Lọc các lệnh an toàn

```python
#đọc các lệnh từ file all_commands.txt và cho dạng pandas
import pandas as pd

# Đọc nội dung của file với bộ mã hóa utf-8
file_path = 'all_commands.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Tạo danh sách các lệnh và giải thích
commands = []
descriptions = []

for line in lines:
    if line.strip():  # Bỏ qua các dòng trống
        # Tách lệnh và giải thích
        parts = line.split('\t', 1)
        if len(parts) == 2:
            command, description = parts
            commands.append(command.strip())
            descriptions.append(description.strip())

# Tạo DataFrame từ danh sách
df = pd.DataFrame({
    'Command': commands,
    'Description': descriptions
})

df
```

```python
# Danh sách chỉ số cần loại bỏ dưới dạng chuỗi
exclude_ranges = "#1-19,60-67,116,113,117,140,160-163,180-185,200-202,204,210,220,236,240,242,262-266,268-270,272,274-280,283,304-307,320-323,325-327,330-333,335,337,342,346,361,362,379,420-424,432,432"

# Xóa ký tự # và chia chuỗi thành các khoảng và số đơn lẻ
exclude_ranges = exclude_ranges.lstrip('#')
exclude_ranges = exclude_ranges.split(',')

# Tạo danh sách chỉ số cần loại bỏ
exclude_indices = set()
for part in exclude_ranges:
    if '-' in part:
        start, end = map(int, part.split('-'))
        exclude_indices.update(range(start, end + 1))
    else:
        exclude_indices.add(int(part))

# Lọc DataFrame để loại bỏ các chỉ số trong danh sách exclude_indices
filtered_df = df.drop(index=exclude_indices, errors='ignore')

filtered_df
```
